````python
# make sure to install selenium and pandas via:
## code: console
## ``` pip install selenium ```
## ``` pip install pandas ```


from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support import select
from selenium.webdriver.support.ui import Select
import pandas as pd

import time
# update your path for the selenium webdriver.
PATH = 'C:\Program Files (x86)\chromedriver.exe'
driver = webdriver.Chrome(PATH)

counter = 0

driver.get('https://registrar.kfupm.edu.sa/courses-classes/course-offering/')

#choosing the term, the value for 212 as shown in the HTML code is '202120'

el = driver.find_element_by_id("course_term_code")
select = Select(el)

select.select_by_value("202120")
el = driver.find_element_by_id("course_dept_code")
select = Select(el)
select.select_by_value("ACFN")
# sleeping so that if internet issues happend, the driver could still reach to the 'Table' HTML tag.
print("Sleeping 1")
time.sleep(5)

# select.deselect_all
table = driver.find_element_by_id('data-table')
# header = table.find_elements_by_tag_name('th')
body = table.find_element_by_tag_name('tbody')

cells = table.find_elements_by_tag_name('td')

#spaghetti code :P, it could've been written in a more efficient way. 

for cell in cells:
    if(counter == 10):
        counter = 0

    if(counter == 0):
        courses.append(cell.text)
        # df['Course-Sec'] = cell.text
    elif(counter == 1):
        activity.append(cell.text)
        # df['Activity'] = cell.text
    elif(counter == 2):
        crn.append(cell.text)
        # df['CRN'] = cell.text
    elif(counter == 3):
        coursenames.append(cell.text)
        # df['Course_name'] = cell.text
    elif(counter == 4):
        instructors.append(cell.text)
        # df['Instructor'] = cell.text
    elif(counter == 5):
        days.append(cell.text)
        # df['Day'] = cell.text
    elif(counter == 6):
        times.append(cell.text)
        # df['Time'] = cell.text
    elif(counter == 7):
        locations.append(cell.text)
        # df['Location'] = cell.text
    elif(counter == 8):
        status.append(cell.text)
        # df['Status'] = cell.text
    elif(counter == 9):
        waitlist.append(cell.text)
        # df['Waitlist'] = cell.text

    counter += 1

d = {
    'Course-Sec': courses,
    'Activity': activity,
    'CRN': crn,
    'Course_name': coursenames,
    'Instructor': instructors,
    'Day': days,
    'Time': times,
    'Location': locations,
    'Status': status,
    'Waitlist': waitlist
}

df = pd.DataFrame(data = d)
# mycourses = courses.text
print("Sleeping 1")
time.sleep(10)
print("Sleeping Final")
display(df)

time.sleep(3)
driver.quit()
````
